```python
# Recoding values to biggest possible categories
def languages2(lang):
    if lang == 'English':
        return 'English2'
    if lang == 'Spanish':
        return 'Other'
    if lang == 'Italian':
        return 'Other'
    if lang == 'Portuguese':
        return 'Other'
    if lang == 'French':
        return 'Other'
    if lang == 'Hindi':
        return 'Other'
    if lang == 'Pashto':
        return 'Other'
    if lang == 'Chinese':
        return 'Other'
    if lang == 'Russian':
        return 'Other'
    if lang == 'Hebrew':
        return 'Other'
    if lang == 'Arabic':
        return 'Other'
    if lang == 'Greek':
        return 'Other'
    if lang == 'German':
        return 'Other'
    if lang == 'Japanese':
        return 'Other'
    if lang == 'Polish':
        return 'Other'
    if lang == 'Vietnamese':
        return 'Other'
    if lang == 'Korean':
        return 'Other'
    if lang == 'Thai':
        return 'Other'
    else: 
        return 'Unknown' 
```

```python
df2['language_converted2'] = df2['language'].apply(languages2)
df2.language_converted2.value_counts(dropna=False)
```
